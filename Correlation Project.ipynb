```python
import pandas as pd 
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjusts the configuration of the plots we will create

#read the data

df = pd.read_csv(r'C:\Users\nadavlah\OneDrive - Intel Corporation\Desktop\movies.csv')




















```

```python
# read data
df.head(10)
```

```python
# look if theres a missing data

for col in df.columns:
    percent_missing = np.mean(df[col]isnull())
    print('{} - {}%'.format(col,percent_missing))
```

```python
df.dtypes
```

```python
# changing data types of columns

df['budget'] = df['budget'].fillna(0).astype('int64')

df['gross'] = df['gross'].fillna(0).astype('int64')
```

```python
pd.set_option('display.max_rows',None)
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False).head(20)
```

```python
# fin correlation
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('budget vs gross earnings')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')
plt.show
```

```python
# plot the budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"blue"}, line_kws={"color":"red"})
```

```python
# checking correlation
df.corr(method='pearson') #pearson,kendall,spearman
```

```python
correlation_matrix = df.corr(method='pearson')

plt.title('correaltion matric for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')

sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
# categorize numericaly string fields to try and fine more correlations

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head(20)
        
        
        
    
    
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

plt.title('correaltion matric for numeric features')
plt.xlabel('movie features')
plt.ylabel('movie features')

sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]
high_correlation
```
